```python
# importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl
```

```python
# ---------------------- Cuadro inutil ------------------------------#
def graficar_entradas(X):
    # Graficar los datos de entrada
    plt.figure()
    plt.scatter(X[:,0], X[:,1])
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Datos de entrada')
    plt.show()

# Solo sirve para saber que son 5 datos y son muy diferentes
# --------------------- Fin Cuadro inutil --------------------------#
```

```python
# --------------------------- Cajita inutil 2 ------------------------------ #
def graficar_resultados(progreso_error):
    # Graficamos el proceso del error
    plt.figure()
    plt.plot(progreso_error)
    plt.xlabel('Número de epochs')
    plt.ylabel('Error de entrenamiento')
    plt.title('Progreso del error de entrenamiento')
    plt.grid()
    plt.show()

# --------------------------- Fin Cajita inutil 2 ------------------------------ #
```

# Perceptron para 2 clases

```python
# cargamos los datos, dos caracteristicas y una etiqueta al final
datos = np.loadtxt('data_perceptron.txt')
X = datos[:,:2]
y = datos[:,2]

# Reordenamos las etiqueas para que constituyan un vector columna
y = datos[:,2].reshape((datos.shape[0],1))
#print(y)
graficar_entradas(X)
```

```python
# Definimos los valores minimos y maximos que puede tomar cada dimensión
dim1_min, dim1_max, dim2_min, dim2_max = 0,1,0,1
num_output = y.shape[1]
#print(num_output)

# definimos las entradas
dim1 = [dim1_min, dim1_max]
dim2 = [dim2_min, dim2_max]
```

```python
# Definimos el perceptron 
perceptron = nl.net.newp([dim1,dim2], num_output)
```

```python
# Entrenamos el perceptron
# PARAMETROS: 
# - lr = learning rate(Tasa de aprendizaje). Dice que tan rapido hago el paso a la minimización
# - epochs = 100 .Cuantas veces recorro todo el conjunto de entrenamiento.
# - show = para que lo muestre cada 20 #
progreso_error = perceptron.train(X, y, epochs = 100, show = 20, lr = 0.03)
graficar_resultados(progreso_error)
```

## Perceptron con mas clases (4 clases)

```python
# En este caso caragamos los datos del archivo data_nn_simple. Cada observacion 
# Consta de 4 valores númericos, dos son caracteristicas
# y los restantes conforman las etiquetas( 2 bits para representar 4 clases)
# Cargamos los datos

datos = np.loadtxt('data_simple_nn.txt')
X = datos[:,0:2]
y = datos[:,2:]
#print(y)
graficar_entradas(X)
```

```python
# Definimos los valores minimos y maximos para cada dimension
# Dos neuronas para la entrada y dos para la salida
dim1_min, dim1_max = X[:,0].min(), X[:,0].max() 
dim2_min, dim2_max = X[:,1].min(), X[:,1].max() 
# definimos las entradas
dim1 = [dim1_min, dim1_max]
dim2 = [dim2_min, dim2_max]
# Definir el numero de neurinas a la salida
num_output = y.shape[1]
```

```python
# Definimos el perceptron
perceptron = nl.net.newp([dim1,dim2], num_output)
```

```python
# Entrenamiento
# Que el error llegue y se solucione o que ne alcance el numero de epoch
progreso_error = perceptron.train(X, y, epochs = 1000, show = 100, lr = 0.03)
graficar_resultados(progreso_error)
```

```python
# Generamos datos de validación
data_test = [[0.4,4.3],[4.4,0.6],[4.7,8.1]]
for item in data_test:
    print(item, '-->',perceptron.sim([item])[0])
```

```python
print(perceptron.sim([[4.4,0.6]]))
```

# ANN multicapa para prediccion

Generamos los datos para la función 
    $$y = 3x^2+5$$

```python
# Generamos el conjunto de entrenamiento
val_min = -15
val_max = 15
num_datos = 130
X = np.linspace(val_min,val_max, num_datos)
y = (3*X**2)+5

# normalizamos y con respecto a la norma 
y /= np.linalg.norm(y)

# Organizar los datos para que la red los recibaa
X = X.reshape(num_datos,1)
y = y.reshape(num_datos,1)
graficar_entradas(X)
```

```python
# Definimos la red neuronal con dos capas ocultas.
# Vamos a tener 3 nueronas en la primera capa y 6 en la segunda capa
# la capa de salida tiene 1 neurona y la de entranda 1 neurona
# Solo tiende a adelante, no se realimenta
ANN = nl.net.newff([[val_min,val_max]],[10,6,1])
ANN.trainf = nl.train.train_gd # especificamos que el algoritmo de optimización será gradiente decendiente
# goal = un error al que quiero llegar
progreso_erro = ANN.train(X, y, epochs = 5000, show = 250, goal = 0.01)
```

```python
# Generamos la predicción
y_pred = ANN.sim(X)
y_pred = y_pred.reshape(num_datos)
graficar_resultados(progreso_error)
```
